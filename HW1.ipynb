Import Data from Yelp Dataset to Python and Elasticsearch

```python
from tqdm import tqdm
import json
```

```python
tables = ['business','checkin','photo','review','tip','user']
for table in tqdm(tables):
    data = []
    _ = es.indices.create(index='yelp_'+table , ignore=400)
    with open('E:\yelp\yelp_academic_dataset_{}.json'.format(table), encoding='utf-8') as f:
        i = 0
        for line in f:
            result=es.create(index='yelp_'+table,doc_type=table,id=i,body=json.loads(line))
            i += 1
            if i % 5000 == 0:
                print(i, 'records have been uploaded')
    print(table, 'data finished!!!')
```

以上程式碼可以成功把數據寫入Elasticsearch, 但寫入速度太慢,用了一天時間才寫入4個Dataset!
